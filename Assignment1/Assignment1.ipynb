```python
import pandas as pd
```

```python
# a) read data from different formats (like csv, xls)

dataset = pd.read_csv("Customers.csv")
dataset.head()
```

```python
# b) Find Shape of Data

dataset.shape
```

```python
# c) Find Missing Values

dataset.isnull().sum()
```

```python
filtered_ds = dataset.dropna()
filtered_ds
```

```python
filtered_ds.shape
```

```python
filtered_ds.isnull().sum()
```

```python
# d) Find data type of each column

filtered_ds.info()
```

```python
# e) Finding out Zero's

(filtered_ds == 0).sum()
```

```python
filtered_ds.loc[:, 'Age'] = filtered_ds['Age'].replace(0, filtered_ds['Age'].mean())
filtered_ds
```

```python
(filtered_ds == 0).sum()
```

```python
# f) Indexing and selecting data, sort data,

filtered_ds.iloc[3:10]
```

```python
sorted_ds = filtered_ds.sort_values(by='Age', ascending=True)
sorted_ds.head()
```

```python
# g) Describe attributes of data, checking data types of each column,

sorted_ds.info()
```

```python
sorted_ds['Age'] = sorted_ds['Age'].astype('int64')
```

```python
sorted_ds.info()
```

```python
# h) counting unique values of data, format of each column, converting variable

sorted_ds.nunique()
```

```python
sorted_ds['Gender'].value_counts()
```

```python
# data type (e.g. from long to short, vice versa)

sorted_ds.dtypes
```

```python
sorted_ds['Work Experience'] = sorted_ds['Work Experience'].astype('float')
```

```python
sorted_ds.dtypes
```
