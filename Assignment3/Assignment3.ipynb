```python
# Importing Libraries

import numpy as np
import pandas as pd
```

```python
# Read the data

df = pd.read_csv("Customers.csv")
```

```python
# Display first 5 documents

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter Plot

plt.figure(figsize=(10, 6))
plt.scatter(x='Annual Income ($)', y='Spending Score (1-100)', data=df, color='red')
plt.title('Annual Icome VS Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
```

```python
# Bar Plot

plt.figure(figsize=(15,6))

sns.barplot(x='Age', y='Annual Income ($)', data=df, color='green')

plt.xticks(rotation=90)
plt.title('Age VS Annaul Income')

plt.show()
```

```python
# Box Plot

plt.figure(figsize=(10, 6))
sns.boxplot(x='Profession', y='Spending Score (1-100)', data=df, color='purple')
plt.xticks(rotation=45)
plt.title('Profession VS Spending Score')
plt.show()
```

```python
# Pie Chart

family_size_counts = df['Profession'].value_counts()
plt.figure(figsize=(8, 8))
family_size_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90, explode=[0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02, ])
plt.title('Family Size Distribution')
# plt.legend()
plt.show()
```

```python
# Line Plot

data_sorted = df.sort_values(by='Age')

plt.figure(figsize=(10, 6))
plt.plot(data_sorted['Age'], data_sorted['Spending Score (1-100)'], marker='o', linestyle='-', color='indigo', label='Spending Score')
plt.title('Spending Score by Age')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.grid()
plt.show()
```
